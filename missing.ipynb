### Imputation Techniques
### 1-Mean Value Imputation



```python
sns.displot(df['age'])
```

```python
df['Mean_Age']=df['age'].fillna(df['age'].mean())
```

```python
df.head()
```

```python
df
```

```python
## This technique work well when your data is normally distributed
```

### 2 -Median Value Imputation
### you have outliers

```python
df['Median_age']=df['age'].fillna(df['age'].median())
```

```python
df
```

### 3-Mode Value Imputation -Categorical

### 3-Mode Value Imputation -Categorical

```python
df[df['embarked'].isnull()]
```

```python
mode=df[df['embarked'].notna()]['embarked'].mode()
```

```python
mode
```

```python
df['new_embarked']=df['embarked'].fillna(mode)
```

```python
df['new_embarked'].isnull()
```
